```python
# 4 — Tokenization Output using CBK Annual Report 2024/25

import PyPDF2
from nltk.tokenize import word_tokenize, sent_tokenize

#  Step 1: Load CBK Annual Report from Google Drive
pdf_path = '/content/1084981846_2025 Annual Report.pdf'

cbk_text = ''
with open(pdf_path, 'rb') as f:
    reader = PyPDF2.PdfReader(f)
    print(f"Total pages in CBK Annual Report: {len(reader.pages)}")

    # Extract pages 7 to 15 (Governor Foreword + Domestic Economy)
    for page_num in range(7, 15):
        page_text = reader.pages[page_num].extract_text()
        if page_text:
            cbk_text += page_text + ' '

print(f"Text extracted: {len(cbk_text):,} characters\n")

#  Step 2: Sentence Tokenization
sentences = sent_tokenize(cbk_text)
print("=" * 55)
print("  4 — TOKENIZATION OUTPUT")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 55)

print(f"\nSENTENCE TOKENIZATION:")
print(f"Total sentences found: {len(sentences)}")
print("\nFirst 5 sentences from the CBK Report:")
for i, sent in enumerate(sentences[:5], 1):
    print(f"\n  Sentence {i}:")
    print(f"  {sent.strip()[:120]}...")

#  Step 3: Word Tokenization
word_tokens = word_tokenize(cbk_text)
print(f"\n{'='*55}")
print(f"WORD TOKENIZATION:")
print(f"Total word tokens : {len(word_tokens):,}")
print(f"First 20 tokens   : {word_tokens[:20]}")
print(f"{'='*55}")
print("Tokenization complete. Text is now ready for preprocessing.")
```

```python
# 5 — Stopwords Removal Output using CBK Annual Report 2024/25

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#  Load stopwords
stop_words = set(stopwords.words('english'))

#  Tokenize and filter
all_tokens = word_tokenize(cbk_text.lower())

# Keep only alphabetic words not in stopwords list
filtered_tokens = [
    word for word in all_tokens
    if word.isalpha() and word not in stop_words
]

#  Results
print("=" * 55)
print("  5 — STOPWORDS REMOVAL OUTPUT")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 55)

print(f"\nBEFORE removal : {len(all_tokens):,} tokens")
print(f"AFTER removal  : {len(filtered_tokens):,} tokens")
print(f"Removed        : {len(all_tokens) - len(filtered_tokens):,} stopwords")
print(f"Reduction      : {((len(all_tokens)-len(filtered_tokens))/len(all_tokens)*100):.1f}%")

print(f"\nSample of REMOVED stopwords (common words with no meaning):")
removed = [w for w in all_tokens if w in stop_words]
print(f"  {list(set(removed))[:15]}")

print(f"\nMeaningful words REMAINING (first 25):")
print(f"  {filtered_tokens[:25]}")

print(f"\nTop 10 most frequent meaningful words in the CBK Report:")
from collections import Counter
freq = Counter(filtered_tokens)
for word, count in freq.most_common(10):
    bar = '█' * count
    print(f"  {word:<20} {count:>3}  {bar}")

print(f"\n{'='*55}")
print("Stopwords removed. Clean tokens ready for NLP analysis.")
```
