```python
# ── SETUP: Load CBK Report for Week 3 Extra Tasks
!pip install PyPDF2 nltk spacy --quiet
!python -m spacy download en_core_web_sm --quiet

import PyPDF2
import nltk
import spacy
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from collections import defaultdict, Counter

nltk.download('punkt',     quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger',     quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

nlp = spacy.load('en_core_web_sm')

# Pages 3–5 = Global Economy, Pages 28–29 = Banking Sector
pdf_path = '/content/1084981846_2025 Annual Report.pdf'

cbk_text = ''
with open(pdf_path, 'rb') as f:
    reader = PyPDF2.PdfReader(f)
    print(f'Total pages: {len(reader.pages)}')
    for page_num in list(range(3, 6)) + list(range(28, 30)):
        page_text = reader.pages[page_num].extract_text()
        if page_text:
            cbk_text += page_text + ' '

print(f'Text loaded: {len(cbk_text):,} characters')
print(f'Preview: {cbk_text[:200]}')
```

```python
# ── EXTRA TASK 1: displaCy NER Visualization
# Course outline Week 3: Information Extraction
# Dataset: CBK Annual Report & Financial Statements 2024/25

from spacy import displacy

# Real paragraph from CBK Annual Report 2024/25 Banking Sector section
cbk_banking = (
    'The Central Bank of Kenya Annual Report 2024/25 was presented by '
    'Governor Dr. Kamau Thugge to the Cabinet Secretary of the National '
    'Treasury in Nairobi. The Monetary Policy Committee lowered the '
    'Central Bank Rate from 13.00 percent in June 2024 to 9.75 percent '
    'by June 2025, to stimulate lending by banks to the private sector. '
    'JPMorgan Chase Bank of the United States was granted authority on '
    'October 14, 2024 to establish a Representative Office in Kenya. '
    'The CBK hosted the 28th MAC meeting in Mombasa on May 9, 2025, '
    'chaired by Dr. Kamau Thugge. The IMF and World Bank praised '
    "Kenya's monetary policy framework. CBK foreign exchange reserves "
    'stood at USD 11082 million as at June 30, 2025. '
    'Overall inflation declined to 3.8 percent in June 2025 '
    'from 4.6 percent in June 2024.'
)

doc = nlp(cbk_banking)

print('=' * 60)
print('  EXTRA TASK 1 — displaCy NER VISUALIZATION')
print('  Dataset: CBK Annual Report & Financial Statements 2024/25')
print('  Pages: 3–5 + 28–29 (Global Economy + Banking Sector)')
print('=' * 60)
print('\nColour-coded NER visualization appears below:')
print('  GREEN  = PERSON    (people names)')
print('  ORANGE = ORG       (organisations)')
print('  BLUE   = GPE       (countries, cities)')
print('  PINK   = DATE      (dates)')
print('  PURPLE = PERCENT   (percentages)')
print('  RED    = MONEY     (money amounts)')
print()

# Render colour-coded entities inline in the notebook
displacy.render(doc, style='ent', jupyter=True)

# Text summary of what was found
print('\n\nEntity Summary:')
entities = defaultdict(list)
for ent in doc.ents:
    entities[ent.label_].append(ent.text)

label_names = {
    'PERSON':'People', 'ORG':'Organisations',
    'GPE':'Countries/Cities', 'DATE':'Dates',
    'PERCENT':'Percentages', 'MONEY':'Money', 'CARDINAL':'Numbers'
}

for label in ['PERSON', 'ORG', 'GPE', 'DATE', 'PERCENT', 'MONEY']:
    ents = entities.get(label, [])
    if ents:
        unique = list(dict.fromkeys(ents))
        print(f'\n  {label_names.get(label, label)} [{label}]:')
        for e in unique:
            print(f'    → {e}')

print('\n' + '=' * 60)
print('WHAT THIS SHOWS:')
print('  displaCy highlights named entities directly in the text.')
print('  Each entity type gets its own colour.')
print('  This is what a real NLP information extraction tool looks like.')
print('  A financial analyst can instantly see all people, orgs,')
print('  dates and figures without reading the full paragraph.')
print('=' * 60)
```

```python
# ── EXTRA TASK 2: Real HMM trained on CBK Report data
# Course outline Week 3 notes Task 4: HMM applications
# Extended to working implementation using hmmlearn
# Dataset: CBK Annual Report & Financial Statements 2024/25

!pip install hmmlearn --quiet

import numpy as np
from hmmlearn import hmm
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# Prepare CBK text as a sequence of numbers
# Each unique word gets a unique ID number
tokens = [
    w.lower() for w in word_tokenize(cbk_text)
    if w.isalpha() and w.lower() not in stop_words and len(w) > 2
]

vocab       = list(set(tokens))
word_to_id  = {w: i for i, w in enumerate(vocab)}
token_ids   = np.array([word_to_id[w] for w in tokens]).reshape(-1, 1)

print('=' * 60)
print('  EXTRA TASK 2 — REAL HMM TRAINED ON CBK REPORT')
print('  Dataset: CBK Annual Report & Financial Statements 2024/25')
print('  Pages: 3–5 + 28–29 (Global Economy + Banking Sector)')
print('=' * 60)
print(f'\nDataset preparation:')
print(f'  Unique words (vocabulary): {len(vocab)}')
print(f'  Total token sequence:      {len(tokens)}')
print(f'  Sample tokens:             {tokens[:10]}')

# Train HMM with 4 hidden states
# 4 states roughly represent 4 types of CBK language:
# financial terms, policy terms, entity terms, number context
n_states = 4

print(f'\nTraining HMM with {n_states} hidden states...')
model = hmm.CategoricalHMM(
    n_components=n_states,
    n_iter=100,
    random_state=42
)
model.fit(token_ids)

print(f'Training complete.')
print(f'  Converged: {model.monitor_.converged}')
print(f'  Iterations run: {model.monitor_.iter}')

# Viterbi decoding — find most likely hidden state sequence
# for the first 20 words of CBK text
sample_tokens = tokens[:20]
sample_ids    = np.array([word_to_id[w] for w in sample_tokens]).reshape(-1, 1)
log_prob, hidden_states = model.decode(sample_ids, algorithm='viterbi')

print(f'\nViterbi Decoding Result:')
print(f'Most likely hidden state sequence for first 20 CBK tokens:\n')
print(f'  {"Word":<22} {"Hidden State":>14}')
print('  ' + '-' * 38)
for word, state in zip(sample_tokens, hidden_states):
    state_label = f'State {state}'
    print(f'  {word:<22} {state_label:>14}')

print(f'\nLog probability of this sequence: {log_prob:.2f}')
print('\n' + '=' * 60)
print('WHAT THIS DEMONSTRATES:')
print('  1. The CBK text was converted to numbers (every word')
print('     gets a unique ID). HMMs work on numbers not text.')
print('  2. The HMM was trained — it learned patterns of which')
print('     words tend to cluster together in CBK language.')
print('  3. Viterbi decoding found the most likely sequence of')
print('     hidden states for the first 20 CBK words.')
print('  4. Words in the same hidden state are similar in how')
print('     they are used in the CBK report.')
print('  Real applications of this same technique:')
print('  → Speech recognition: hidden states = phonemes')
print('  → POS taggers: hidden states = grammar tags')
print('  → Spam detection: hidden states = spam/not-spam patterns')
print('=' * 60)
```
