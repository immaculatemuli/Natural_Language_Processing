```python
# ── WEEK 4 PRACTICAL TASK 1: Dependency Parsing ─
# STEP 1: Install and load spaCy

!pip install spacy --quiet
!python -m spacy download en_core_web_sm --quiet

import spacy
nlp = spacy.load("en_core_web_sm")


print("=" * 55)
print("  Dependency Parsing")
print("=" * 55)

text = "The lecturer teaches Natural Language Processing."
doc  = nlp(text)

print(f"\nSentence: {text}\n")
print(f"{'Word':<20} {'Dependency':<15} {'Head Word'}")
print("-" * 50)
for token in doc:
    print(f"{token.text:<20} {token.dep_:<15} {token.head.text}")

# What to observe
print("\nOBSERVATIONS:")
subject = [t.text for t in doc if t.dep_ == "nsubj"]
root    = [t.text for t in doc if t.dep_ == "ROOT"]
obj     = [t.text for t in doc if t.dep_ in ("dobj", "pobj")]
print(f"  Subject (who does the action) : {subject}")
print(f"  Root Verb (main verb)         : {root}")
print(f"  Object (what receives action) : {obj}")
print(f"\n  Root word = '{root[0] if root else '?'}' — everything in the sentence connects back to this word.")

```

```python
#  WEEK 4 PRACTICAL TASK 2: Semantic Similarity

import spacy
nlp = spacy.load("en_core_web_sm")


print("=" * 55)
print(" Semantic Similarity")
print("=" * 55)

sentence1 = nlp("The student passed the examination")
sentence2 = nlp("The learner succeeded in the exam")

similarity = sentence1.similarity(sentence2)
print(f"\nSentence 1 : {sentence1.text}")
print(f"Sentence 2 : {sentence2.text}")
print(f"Similarity Score: {similarity:.4f}")

print("\nOBSERVATIONS:")
print(f"  Score of {similarity:.4f} means the sentences are very similar in meaning.")
print(f"  'student' and 'learner' mean the same thing.")
print(f"  'passed' and 'succeeded' carry the same outcome.")
print(f"  'examination' and 'exam' are the same word — one is shorter.")
print(f"  The machine understood all this without being told explicitly.")



print("\n" + "=" * 55)
print("Semantic similarity used in: search engines, chatbots,")
print("recommendation systems, and AI assistants.")
```

```python
# ── WEEK 4 CLASS ASSIGNMENT 1: Dependency Structure ─
# Analyze the two sentences

import spacy
nlp = spacy.load("en_core_web_sm")

sentences = [
    "The administrator updated student records.",
    "Machine learning improves language processing.",
]

dep_meanings = {
    "nsubj":   "Subject",
    "ROOT":    "Main Verb (Root)",
    "dobj":    "Direct Object",
    "det":     "Determiner (the/a)",
    "amod":    "Adjective Modifier",
    "compound":"Compound Noun Part",
    "nmod":    "Noun Modifier",
    "nummod":  "Number Modifier",
}

print("=" * 60)
print("  CLASS ASSIGNMENT 1 — DEPENDENCY STRUCTURE ANALYSIS")
print("=" * 60)

for i, sentence in enumerate(sentences, 1):
    doc = nlp(sentence)
    print(f"\nSentence {i}: {sentence}")
    print(f"  {'Word':<25} {'Dependency':<15} {'Head Word'}")
    print(f"  {'-'*55}")
    for token in doc:
        print(f"  {token.text:<25} {token.dep_:<15} {token.head.text}")

    subject = [t.text for t in doc if t.dep_ == "nsubj"]
    root    = [t.text for t in doc if t.dep_ == "ROOT"]
    obj      = [t.text for t in doc if t.dep_ in ("dobj", "attr")]
    print(f"\n  Subject   : {subject}")
    print(f"  Root Verb : {root}")
    print(f"  Object    : {obj}")

print("\n" + "=" * 60)
print("EXPLANATION OF DEPENDENCIES:")
print()
print("Sentence 1: 'The administrator updated student records'")
print("  'administrator' is the subject — the person doing the action")
print("  'updated' is the ROOT — the main verb everything connects to")
print("  'records' is the direct object — what was updated")
print("  'student' is a compound — joins with 'records' to form one noun")
print("  'The' is a determiner — identifies which administrator")
print()
print("Sentence 2: 'Machine learning improves language processing'")
print("  'learning' is the subject — Machine learning does the action")
print("  'improves' is the ROOT — the main verb")
print("  'processing' is the direct object — what is improved")
print("  'Machine' is compound — joins 'learning' to form 'Machine learning'")
print("  'language' is compound — joins 'processing' to form 'language processing'")
print()
print("KEY INSIGHT: Dependency parsing automatically identifies")
print("these relationships without any manual reading.")
print("This is how AI systems understand sentence meaning.")
```

```python
# WEEK 4 CLASS ASSIGNMENT 2: Semantic Similarity Pairs
# Two similar sentences + two unrelated sentences
# Explain which pair has higher similarity and why

import spacy
nlp = spacy.load("en_core_web_sm")

print("=" * 65)
print("  CLASS ASSIGNMENT 2 — SIMILAR vs UNRELATED SENTENCES")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 65)

#  TWO SIMILAR SENTENCE PAIRS (expect high score)
similar = [
    (
        "The CBK reduced the Central Bank Rate to support economic growth.",
        "The central bank cut interest rates to boost the economy.",
        "Similar Pair 1"
    ),
    (
        "Kenya's foreign exchange reserves reached 4.86 months of import cover.",
        "The CBK maintained dollar reserves covering nearly 5 months of imports.",
        "Similar Pair 2"
    ),
]

#  TWO UNRELATED SENTENCE PAIRS (expect low score)
unrelated = [
    (
        "The Monetary Policy Committee approved new banking regulations.",
        "Students should submit their assignments before the deadline.",
        "Unrelated Pair 1"
    ),
    (
        "Inflation declined due to lower food and energy prices in Kenya.",
        "The football match ended in a draw after ninety minutes.",
        "Unrelated Pair 2"
    ),
]

print("\n--- SIMILAR SENTENCE PAIRS ---")
for s1_text, s2_text, label in similar:
    score = nlp(s1_text).similarity(nlp(s2_text))
    print(f"\n[{label}]")
    print(f"  Sentence A : {s1_text}")
    print(f"  Sentence B : {s2_text}")
    print(f"  Score      : {score:.4f}")

print("\n--- UNRELATED SENTENCE PAIRS ---")
for s1_text, s2_text, label in unrelated:
    score = nlp(s1_text).similarity(nlp(s2_text))
    print(f"\n[{label}]")
    print(f"  Sentence A : {s1_text}")
    print(f"  Sentence B : {s2_text}")
    print(f"  Score      : {score:.4f}")

print("\n" + "=" * 65)
print("EXPLANATION — WHICH PAIR HAS HIGHER SIMILARITY AND WHY:")
print()
print("Similar pairs score HIGHER because:")
print("  → They discuss the same concept using different words")
print("  → 'reduced' and 'cut' mean the same thing")
print("  → 'interest rates' and 'borrowing costs' are the same concept")
print("  → 'boost' and 'support' carry similar meanings")
print("  → The machine learned these word relationships from training data")
print()
print("Unrelated pairs score LOWER because:")
print("  → The vocabulary is completely different")
print("  → Financial words (monetary, inflation, CBK) do not appear")
print("    in the academic or sports sentences")
print("  → No shared concepts or word meanings exist between the pairs")
print()
print("REAL WORLD APPLICATION:")
print("  Search engines use semantic similarity to return relevant results")
print("  even when your search words do not exactly match the page content.")
print("  Google understands 'cheap flights' and 'affordable airfare'")
print("  mean the same thing — same principle as this assignment.")
```
