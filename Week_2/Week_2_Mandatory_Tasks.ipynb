```python
# 1 — Bigram and Trigram Output
# Dataset: CBK Annual Report & Financial Statements 2024/25

import PyPDF2
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

# Load CBK PDF report
pdf_path = '/content/1084981846_2025 Annual Report.pdf'

cbk_text = ''
with open(pdf_path, 'rb') as f:
    reader = PyPDF2.PdfReader(f)
    # Pages 22–27 = Monetary Operations + Interest Rates section
    for page_num in range(22, 27):
        page_text = reader.pages[page_num].extract_text()
        if page_text:
            cbk_text += page_text + ' '

print(f"CBK Report text loaded: {len(cbk_text):,} characters\n")

# Clean tokens (keep all words for N-grams)
tokens = [w.lower() for w in word_tokenize(cbk_text) if w.isalpha()]

#  Build Bigrams
bigram_list  = list(ngrams(tokens, 2))
bigram_freq  = Counter(bigram_list)

#  Build Trigrams
trigram_list = list(ngrams(tokens, 3))
trigram_freq = Counter(trigram_list)

print("=" * 55)
print("  1 — BIGRAM AND TRIGRAM OUTPUT")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 55)

print("\nTOP 10 BIGRAMS (2-word phrases):")
print("-" * 40)
for bg, count in bigram_freq.most_common(10):
    print(f"  {' '.join(bg):<35} {count}x")

print("\nTOP 10 TRIGRAMS (3-word phrases):")
print("-" * 40)
for tg, count in trigram_freq.most_common(10):
    print(f"  {' '.join(tg):<45} {count}x")

print("\n" + "=" * 55)
print("N-grams reveal the dominant phrases in the CBK Report.")
print("'monetary policy' and 'central bank' should appear most.")
```

```python
# 2 — POS Tagging Example
# Using a real sentence from the CBK Annual Report 2024/25

import nltk
nltk.download('averaged_perceptron_tagger', quiet=True)

# Real sentence from CBK Annual Report — Monetary Policy section
cbk_sentence = (
    "The Monetary Policy Committee lowered the Central Bank Rate "
    "from 13.00 percent in June 2024 to 9.75 percent by June 2025."
)

# Tokenize and tag
tokens = nltk.word_tokenize(cbk_sentence)
pos_tags = nltk.pos_tag(tokens)

# Tag meanings for easy reading
tag_meanings = {
    'DT':'Determiner',   'NN':'Noun',          'NNS':'Noun(plural)',
    'NNP':'Proper Noun', 'VBD':'Verb(past)',    'VBZ':'Verb(present)',
    'VB':'Verb',         'VBN':'Verb(past part)','JJ':'Adjective',
    'RB':'Adverb',       'IN':'Preposition',    'CD':'Number',
    'TO':'To',           'CC':'Conjunction',    'PRP':'Pronoun',
}

print("=" * 55)
print("  2 — POS TAGGING EXAMPLE")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 55)
print(f"\nSentence:\n  \"{cbk_sentence}\"\n")
print(f"{'Word':<25} {'POS Tag':<12} {'Meaning'}")
print("-" * 55)
for word, tag in pos_tags:
    meaning = tag_meanings.get(tag, tag)
    print(f"  {word:<23} {tag:<12} {meaning}")

print("\n" + "=" * 55)
print("Every word in the CBK sentence has been labelled.")
print("Proper Nouns (NNP) mark: Monetary, Policy, Committee,")
print("Central, Bank, June — key entities in the report.")
```

```python
# 3 — Python Code for POS Tagging
# Tagging multiple CBK sentences to show the full code workflow

import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab', quiet=True) # Download punkt_tab resource
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

# Three sentences taken directly from CBK Annual Report 2024/25
cbk_sentences = [
    "Overall inflation declined to 3.8 percent in June 2025.",
    "The banking sector remained stable and resilient.",
    "Governor Dr. Kamau Thugge announced key reforms during the year."
]

print("=" * 55)
print("  3 — PYTHON CODE FOR POS TAGGING")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 55)

for i, sentence in enumerate(cbk_sentences, 1):
    print(f"\nSentence {i}: {sentence}")

    # Step 1: Tokenize
    tokens = word_tokenize(sentence)

    # Step 2: POS tag
    tagged = nltk.pos_tag(tokens)

    # Step 3: Print result
    print(f"POS Tags: {tagged}")

print("\n" + "=" * 55)
print("CODE STEPS:")
print("1. word_tokenize()  → splits sentence into words")
print("2. nltk.pos_tag()   → labels each word with its grammar role")
print("3. Output           → list of (word, tag) pairs")
print("=" * 55)
```

```python
# 4 — Language Prediction Example
# Building a simple next-word predictor from CBK Report text

from collections import defaultdict, Counter
from nltk.tokenize import word_tokenize

# Use the same cbk_text loaded in Cell 1
# (make sure Cell 1 ran first)
tokens = [w.lower() for w in word_tokenize(cbk_text) if w.isalpha()]

# Build prediction model using bigrams
# For each word, record what words follow it
predict_model = defaultdict(Counter)
for i in range(len(tokens) - 1):
    predict_model[tokens[i]][tokens[i + 1]] += 1

def predict_next(word, top_n=3):
    word = word.lower()
    options = predict_model.get(word, {})
    if not options:
        return ["(word not found in CBK report)"]
    return [w for w, _ in Counter(options).most_common(top_n)]

# Test with key CBK financial terms
test_words = [
    'monetary',
    'banking',
    'inflation',
    'central',
    'lending',
    'policy',
    'exchange',
    'percent',
]

print("=" * 55)
print("  4 — LANGUAGE PREDICTION EXAMPLE")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 55)
print("\nBased on patterns learned from the CBK Report,")
print("what word most likely follows each keyword?\n")
print(f"{'Keyword':<18} {'Top 3 Predicted Next Words'}")
print("-" * 55)
for word in test_words:
    predictions = predict_next(word)
    print(f"  {word:<16} →  {predictions}")

print("\n" + "=" * 55)
print("This is how autocomplete and basic chatbots work.")
print("The model learned from CBK text — so 'monetary'")
print("should predict 'policy', 'central' should predict 'bank'.")
print("=" * 55)
```

```python
# 5 — Sentence Analysis Output
# Full sentence-level analysis of the CBK Report

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter

# Use same cbk_text from Cell 1
sentences = sent_tokenize(cbk_text)

tag_meanings = {
    'DT':'Determiner',   'NN':'Noun',       'NNS':'Noun(plural)',
    'NNP':'Proper Noun', 'VBD':'Verb(past)','JJ':'Adjective',
    'IN':'Preposition',  'CD':'Number',     'RB':'Adverb',
    'VBZ':'Verb(present)','CC':'Conjunction',
}

print("=" * 60)
print("  5 — SENTENCE ANALYSIS OUTPUT")
print("  Dataset: CBK Annual Report & Financial Statements 2024/25")
print("=" * 60)

# Analyse first 4 sentences from the CBK section
for i, sent in enumerate(sentences[:4], 1):
    tokens  = word_tokenize(sent)
    tagged  = nltk.pos_tag(tokens)

    # Count POS tag types in this sentence
    tag_counts = Counter(tag for _, tag in tagged)
    nouns  = tag_counts.get('NN', 0) + tag_counts.get('NNP', 0) + tag_counts.get('NNS', 0)
    verbs  = sum(v for k, v in tag_counts.items() if k.startswith('VB'))
    adjs   = tag_counts.get('JJ', 0)
    nums   = tag_counts.get('CD', 0)

    print(f"\nSentence {i}:")
    print(f"  Text   : {sent.strip()[:100]}...")
    print(f"  Length : {len(tokens)} words")
    print(f"  Nouns  : {nouns}  |  Verbs: {verbs}  |  Adjectives: {adjs}  |  Numbers: {nums}")
    print(f"  Tags   : {tagged[:8]}...")

# Overall statistics
all_tokens = word_tokenize(cbk_text)
all_tagged = nltk.pos_tag([w for w in all_tokens if w.isalpha()])
all_tag_counts = Counter(tag for _, tag in all_tagged)

print("\n" + "=" * 60)
print("OVERALL SENTENCE ANALYSIS — CBK Report Section:")
print(f"  Total sentences    : {len(sentences)}")
print(f"  Total words        : {len(all_tokens):,}")
print(f"  Total Proper Nouns : {all_tag_counts.get('NNP', 0)} (names, places, orgs)")
print(f"  Total Numbers (CD) : {all_tag_counts.get('CD', 0)} (figures, percentages)")
print(f"  Total Verbs        : {sum(v for k,v in all_tag_counts.items() if k.startswith('VB'))}")
print(f"  Total Adjectives   : {all_tag_counts.get('JJ', 0)}")
print("=" * 60)
print("High proper noun count = many named entities in CBK text.")
print("High number count = lots of financial figures — ideal for NLP.")
```
