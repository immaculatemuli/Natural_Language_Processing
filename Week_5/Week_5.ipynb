```python
# WEEK 5 MINI PROJECT
# Project Title: Student Academic Assistant Chatbot
# Unit: BIT4133 - Natural Language Processing with Deep Learning
#
# PROJECT REQUIREMENTS:
# ✓ Responds to greetings
# ✓ Answers CAT schedule questions
# ✓ Answers unit registration questions
# ✓ Exits correctly
#
# NLP CONCEPTS IMPLEMENTED (Weeks 1-5):
# Week 1 → Tokenization, Stop Words Removal, Stemming, Lemmatization
# Week 2 → N-gram Model, POS Tagging
# Week 3 → Sequence Labeling (HMM concept)
# Week 4 → Dependency Parsing, Semantic Similarity
# Week 5 → Full NLP Pipeline integrated into chatbot

#  INSTALL AND IMPORT ALL LIBRARIES
!pip install spacy --quiet
!python -m spacy download en_core_web_sm --quiet

import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.util import ngrams
from collections import Counter

# Download all required NLTK data
nltk.download('punkt',                          quiet=True)
nltk.download('punkt_tab',                      quiet=True)
nltk.download('stopwords',                      quiet=True)
nltk.download('wordnet',                        quiet=True)
nltk.download('averaged_perceptron_tagger',     quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

# Load spaCy model for parsing and semantic similarity
nlp_spacy = spacy.load('en_core_web_sm')

# Initialize NLP tools
stemmer    = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

print("=" * 60)
print("  All NLP libraries loaded successfully")
print("  Week 1-5 tools ready: NLTK + spaCy")
print("=" * 60)



# STEP 1: PREPROCESSING PIPELINE (Week 1)
# Cleans user input using all Week 1 techniques


def preprocess_input(user_input):
    """
    Applies full Week 1 NLP preprocessing pipeline to user input.
    Same techniques used on CBK Annual Report in weekly tasks.
    """

    #  Lowercase conversion
    cleaned = user_input.lower().strip()

    #  Tokenization
    # Same word_tokenize() used on CBK report sentences
    tokens = word_tokenize(cleaned)

    #  Stop Words Removal
    # Same stop_words removal used on CBK Governor Foreword
    filtered_tokens = [
        w for w in tokens
        if w.isalpha() and w not in stop_words
    ]

    #  Stemming
    # Same PorterStemmer used on CBK financial words
    # e.g. "registering" → "regist", "examining" → "examin"
    stemmed_tokens = [stemmer.stem(w) for w in filtered_tokens]

    #  Lemmatization
    # Same WordNetLemmatizer used on CBK report vocabulary
    # e.g. "examinations" → "examination", "fees" → "fee"
    lemmatized_tokens = [lemmatizer.lemmatize(w) for w in filtered_tokens]

    return {
        'cleaned':    cleaned,
        'tokens':     tokens,
        'filtered':   filtered_tokens,
        'stemmed':    stemmed_tokens,
        'lemmatized': lemmatized_tokens,
    }


# STEP 2: N-GRAM MODEL (Week 2)
# Builds bigrams from user input for phrase detection

def extract_ngrams(tokens):
    """
    Extracts bigrams and trigrams from user input tokens.
    """
    bigrams  = list(ngrams(tokens, 2))
    trigrams = list(ngrams(tokens, 3))

    bigram_phrases  = [' '.join(bg) for bg in bigrams]
    trigram_phrases = [' '.join(tg) for tg in trigrams]

    return bigram_phrases, trigram_phrases


# STEP 3: POS TAGGING
# Labels each word in user input with its grammatical role


def pos_tag_input(tokens):
    """
    Tags each token with its Part of Speech.
    Identifies verbs (actions) and nouns (topics) in user input.
    Same technique used on CBK interest rate sentences.
    """
    pos_tags = nltk.pos_tag(tokens)

    # Extract action words (verbs) and topic words (nouns)
    verbs = [w for w, t in pos_tags if t.startswith('VB')]
    nouns = [w for w, t in pos_tags if t.startswith('NN')]

    return pos_tags, verbs, nouns


# STEP 4: SEQUENCE LABELING
# Labels each token as COMMAND, TOPIC, or OTHER
# Mimics how HMM sequence labeling assigns labels to words

def label_sequence(tokens):
    """
    Assigns a sequence label to every token in user input.
    Demonstrates HMM sequence labeling concept from Week 3.
    COMMAND = action words, TOPIC = subject words, OTHER = rest.
    """
    # Words that represent ACTIONS
    command_words = {
        'add', 'drop', 'register', 'enroll', 'join',
        'submit', 'check', 'revise', 'prepare', 'help',
        'tell', 'show', 'explain', 'know', 'find', 'get'
    }

    # Words that represent TOPICS
    topic_words = {
        'cat', 'exam', 'test', 'assessment', 'unit',
        'fee', 'fees', 'registration', 'logbook', 'github',
        'bit4133', 'assignment', 'marks', 'grade', 'course',
        'schedule', 'timetable', 'deadline', 'certificate',
        'chatbot', 'nlp', 'python', 'colab', 'notebook'
    }

    labeled_sequence = []
    for token in tokens:
        if token.lower() in command_words:
            label = 'COMMAND'
        elif token.lower() in topic_words:
            label = 'TOPIC'
        else:
            label = 'OTHER'
        labeled_sequence.append((token, label))

    return labeled_sequence

# STEP 5: DEPENDENCY PARSING
# Identifies subject, verb, object in user input


def parse_dependencies(user_input):
    """
    Parses grammatical dependencies in user input.
    Finds what the user is doing (ROOT verb) and about what (object).
    Same token.dep_ technique used on CBK policy sentences.
    """
    doc = nlp_spacy(user_input)

    root    = [t.text for t in doc if t.dep_ == 'ROOT']
    subject = [t.text for t in doc if t.dep_ == 'nsubj']
    obj     = [t.text for t in doc if t.dep_ in ('dobj', 'pobj', 'attr')]

    return root, subject, obj


# STEP 6: SEMANTIC SIMILARITY FALLBACK
# When no keyword matches, find the closest meaning

# Reference phrases the chatbot understands — with their category
reference_phrases = {
    "hello how are you":           "greeting",
    "good morning":                "greeting",
    "cat exam schedule":           "cat_general",
    "cat one topics":              "cat1",
    "cat two requirements":        "cat2",
    "how to add a unit":           "add_unit",
    "how to drop a unit":          "drop_unit",
    "unit registration process":   "registration",
    "goodbye see you later":       "exit",
}

def semantic_fallback(user_input):
    """
    Uses semantic similarity to understand input when keywords fail.
    Same spaCy similarity() used on CBK sentence pairs in Week 4.
    Scores above 0.75 are considered a match.
    """
    user_doc   = nlp_spacy(user_input.lower())
    best_match = None
    best_score = 0.0

    for phrase, category in reference_phrases.items():
        ref_doc = nlp_spacy(phrase)
        score   = user_doc.similarity(ref_doc)
        if score > best_score:
            best_score = score
            best_match = category

    # Only return a match if similarity is high enough
    if best_score > 0.75:
        return best_match, best_score
    return None, best_score


# STEP 7: RESPONSE GENERATION
# Maps detected intent to helpful responses

def get_response(cleaned, bigram_phrases, lemmatized_tokens, sequence_labels):
    """
    Determines chatbot response based on NLP analysis of input.
    Uses bigrams (Week 2), lemmatized tokens (Week 1),
    and sequence labels (Week 3) to find intent.
    """

    # Collect TOPIC and COMMAND labels from sequence (Week 3)
    topics   = [w for w, label in sequence_labels if label == 'TOPIC']
    commands = [w for w, label in sequence_labels if label == 'COMMAND']

    #  GREETINGS
    greeting_words = {'hello', 'hi', 'hey', 'morning', 'afternoon',
                      'evening', 'greetings', 'welcome'}
    if any(w in greeting_words for w in lemmatized_tokens) or \
       any(w in greeting_words for w in cleaned.split()):
        return (
            "Hello! Welcome to the BIT4133 Student Assistant.\n"
            "I can help you with:\n"
            "  → CAT 1 and CAT 2 schedule questions\n"
            "  → Unit registration (add or drop)\n"
            "  → NLP revision topics\n"
            "  → Logbook and GitHub guidance\n"
            "What do you need help with today?"
        ), False, "GREETING"

    #  CAT 1
    # Check bigrams first
    if 'cat 1' in bigram_phrases or \
       ('cat' in topics and '1' in cleaned) or \
       'cat1' in cleaned:
        return (
            "CAT 1 covers Weeks 1 to 5:\n"
            "  Week 1 → Tokenization, stop words, stemming, lemmatization\n"
            "  Week 2 → N-gram models and POS tagging\n"
            "  Week 3 → HMMs, sequence labeling, NER\n"
            "  Week 4 → Dependency parsing, semantic similarity\n"
            "  Week 5 → Full NLP pipeline and chatbot\n\n"
            "Make sure your logbook has:\n"
            "  ✓ All figures completed with screenshots\n"
            "  ✓ 100-word reflection per week\n"
            "  ✓ Notebooks pushed to GitHub"
        ), False, "CAT1"

    #  CAT 2
    if 'cat 2' in bigram_phrases or \
       ('cat' in topics and '2' in cleaned) or \
       'cat2' in cleaned:
        return (
            "CAT 2 requires ONE free online NLP or AI course.\n\n"
            "You must submit:\n"
            "  → Certificate or completion screenshot\n"
            "  → 1 to 2 page reflection report\n"
            "  → Key skills learned\n"
            "  → How it connects to BIT4133\n\n"
            "Recommended free courses:\n"
            "  1. Hugging Face NLP Course → huggingface.co/learn\n"
            "  2. Google AI Course        → grow.google/ai\n"
            "  3. IBM SkillsBuild         → skillsbuild.org\n"
            "  4. OpenAI Academy          → academy.openai.com\n\n"
            "Start early — courses take time to complete."
        ), False, "CAT2"

    #  GENERAL CAT / EXAM / REVISION
    cat_words = {'cat', 'exam', 'test', 'assessment',
                 'revision', 'revise', 'prepare', 'mark', 'grade'}
    if any(w in cat_words for w in lemmatized_tokens) or \
       any(w in cat_words for w in topics):
        return (
            "BIT4133 has two assessments:\n\n"
            "  CAT 1 → Practical logbook covering Weeks 1 to 5\n"
            "          Tested: tokenization, POS tagging, N-grams,\n"
            "          HMMs, parsing, and semantic similarity\n\n"
            "  CAT 2 → Free online NLP course certificate\n"
            "          + 1 to 2 page reflection report\n\n"
            "Type 'cat 1' for CAT 1 details.\n"
            "Type 'cat 2' for CAT 2 details."
        ), False, "CAT_GENERAL"

    #  ADD UNIT
    # Bigram detection (Week 2) catches "add unit" as a phrase
    if 'add unit' in bigram_phrases or \
       'enrol unit' in bigram_phrases or \
       ('add' in commands and 'unit' in topics) or \
       any(w in {'enroll', 'enrol', 'join'} for w in lemmatized_tokens):
        return (
            "To ADD a unit:\n"
            "  → Visit the registrar office\n"
            "  → Bring your student ID and admission number\n"
            "  → Request a unit addition form\n"
            "  → Submit BEFORE the add/drop deadline\n"
            "  → Check the student portal for the deadline date\n\n"
            "Note: Late additions may not be accepted."
        ), False, "ADD_UNIT"

    #  DROP UNIT
    if 'drop unit' in bigram_phrases or \
       'withdraw unit' in bigram_phrases or \
       ('drop' in commands and 'unit' in topics) or \
       any(w in {'withdraw', 'remove'} for w in lemmatized_tokens):
        return (
            "To DROP a unit:\n"
            "  → Visit the registrar office\n"
            "  → Request a unit drop form\n"
            "  → Fill and submit before the deadline\n"
            "  → Bring your student ID\n"
            "  → Confirm whether fees are affected\n\n"
            "Note: Dropping a unit may affect your fee balance."
        ), False, "DROP_UNIT"

    #  GENERAL REGISTRATION
    registration_words = {'register', 'registration', 'unit', 'course',
                          'fee', 'portal', 'timetable', 'admission',
                          'bit4133', 'bit', 'nlp'}
    if any(w in registration_words for w in lemmatized_tokens) or \
       any(w in registration_words for w in topics):
        return (
            "For registration and unit queries:\n\n"
            "  Registrar office → unit add, drop, transfers\n"
            "  Finance office   → fees and payments\n"
            "  Student portal   → timetables and results\n\n"
            "BIT4133 details:\n"
            "  Unit  : Natural Language Processing with Deep Learning\n"
            "  Tools : Python, NLTK, spaCy, Google Colab, GitHub\n"
            "Type 'add unit' or 'drop unit' for specific steps."
        ), False, "REGISTRATION"

    #  EXIT
    exit_words = {'bye', 'goodbye', 'exit', 'quit', 'done', 'thank', 'thanks'}
    if any(w in exit_words for w in lemmatized_tokens) or \
       any(w in exit_words for w in cleaned.split()):
        return (
            "Goodbye! All the best with BIT4133.\n\n"
            "Before you go, remember to:\n"
            "  ✓ Complete your logbook figures with screenshots\n"
            "  ✓ Write 100-word reflections for each week\n"
            "  ✓ Push all notebooks to GitHub\n"
            "  ✓ Start your free online course for CAT 2\n\n"
            "Good luck with your NLP journey!"
        ), True, "EXIT"

    #  DEFAULT (no keyword matched)
    return (
        "I did not understand that.\n"
        "I can help with:\n"
        "  → Greetings         (try: 'hello')\n"
        "  → CAT 1 questions   (try: 'cat 1')\n"
        "  → CAT 2 questions   (try: 'cat 2')\n"
        "  → Add a unit        (try: 'add unit')\n"
        "  → Drop a unit       (try: 'drop unit')\n"
        "  → Registration help (try: 'register')\n"
        "  → Exit              (try: 'bye')"
    ), False, "DEFAULT"


# STEP 8: MAIN CHATBOT LOOP
# Full NLP pipeline — input → preprocess → analyze → respond

print("\n")
print("=" * 60)
print("  WEEK 5 MINI PROJECT")
print("  Student Academic Assistant Chatbot")
print("  BIT4133 — Natural Language Processing")
print("  NLP Pipeline: Weeks 1 to 5 integrated")
print("=" * 60)
print("\nWelcome! I am your BIT4133 Student Assistant.")
print("Type 'bye' to exit.\n")
print("-" * 60)

interaction_count = 0

while True:

    #  GET USER INPUT
    user_input = input("\nYou: ")

    # Handle empty input
    if not user_input.strip():
        print("Bot: Please type something. I am here to help!")
        continue

    interaction_count += 1

    print("\n[NLP PIPELINE OUTPUT]")
    print("-" * 40)

    #  WEEK 1: PREPROCESSING
    processed = preprocess_input(user_input)
    print(f"[Week 1] Tokens          : {processed['tokens']}")
    print(f"[Week 1] After stopwords : {processed['filtered']}")
    print(f"[Week 1] Stemmed         : {processed['stemmed']}")
    print(f"[Week 1] Lemmatized      : {processed['lemmatized']}")

    # WEEK 2: N-GRAMS
    bigram_phrases, trigram_phrases = extract_ngrams(processed['tokens'])
    print(f"[Week 2] Bigrams         : {bigram_phrases}")

    # WEEK 2: POS TAGGING
    pos_tags, verbs, nouns = pos_tag_input(processed['tokens'])
    print(f"[Week 2] POS Tags        : {pos_tags}")
    print(f"[Week 2] Verbs           : {verbs}")
    print(f"[Week 2] Nouns           : {nouns}")

    # WEEK 3: SEQUENCE LABELING
    sequence_labels = label_sequence(processed['tokens'])
    print(f"[Week 3] Sequence Labels : {sequence_labels}")

    #  WEEK 4: DEPENDENCY PARSING
    root, subject, obj = parse_dependencies(user_input)
    print(f"[Week 4] Root Verb       : {root}")
    print(f"[Week 4] Subject         : {subject}")
    print(f"[Week 4] Object          : {obj}")

    #  GET RESPONSE
    response, should_exit, intent = get_response(
        processed['cleaned'],
        bigram_phrases,
        processed['lemmatized'],
        sequence_labels
    )

    print(f"[Intent Detected]        : {intent}")

    # WEEK 4: SEMANTIC FALLBACK
    # Only runs if intent was DEFAULT (no keyword matched)
    if intent == "DEFAULT":
        fallback_intent, sim_score = semantic_fallback(user_input)
        print(f"[Week 4] Semantic Score  : {sim_score:.3f}")
        if fallback_intent:
            print(f"[Week 4] Semantic Match  : {fallback_intent}")

    print("-" * 40)

    # PRINT RESPONSE
    print(f"\nBot: {response}\n")

    #  EXIT CHECK
    if should_exit:
        print(f"\n[Session ended after {interaction_count} interactions]")
        break


# ════════════════════════════════════════════════════════════════
# DELIVERABLE 3: SHORT EXPLANATION DOCUMENT
# ════════════════════════════════════════════════════════════════
#
# PROJECT TITLE: Student Academic Assistant Chatbot
# UNIT: BIT4133 — Natural Language Processing with Deep Learning
#
# WHAT THE CHATBOT DOES:
# The Student Academic Assistant is a rule-based chatbot that
# applies every NLP concept taught across Weeks 1 to 5 of BIT4133.
# It understands student questions about CAT schedules, unit
# registration, and greetings, then returns relevant responses.
#
# HOW IT WORKS — STEP BY STEP:
#
# Step 1 — User types a message (e.g. "I want to add a unit")
#
# Step 2 — Week 1 preprocessing runs:
#   .lower()              → converts to lowercase
#   word_tokenize()       → splits into words
#   stop_words removal    → removes 'I', 'to', 'a'
#   PorterStemmer         → stems remaining words
#   WordNetLemmatizer     → lemmatizes remaining words
#
# Step 3 — Week 2 N-gram model runs:
#   ngrams(tokens, 2)     → builds bigrams
#   "add unit" detected   → precise phrase matching
#
# Step 4 — Week 2 POS tagging runs:
#   nltk.pos_tag()        → labels each word
#   'add' tagged as VB    → it is an action word
#   'unit' tagged as NN   → it is a topic word
#
# Step 5 — Week 3 sequence labeling runs:
#   'add' → COMMAND label
#   'unit' → TOPIC label
#   Together → confirms intent is ADD_UNIT
#
# Step 6 — Week 4 dependency parsing runs:
#   ROOT verb = 'add'     → confirms the action
#   object = 'unit'       → confirms what the action is about
#
# Step 7 — Response matched and printed
#
# Step 8 — If no keyword matched, Week 4 semantic similarity
#   runs as a fallback using spaCy sentence.similarity()
#
# WHAT THE CHATBOT COVERS:
# ✓ Greetings    → hello, hi, hey, good morning, good afternoon
# ✓ CAT 1        → cat 1, week topics, logbook requirements
# ✓ CAT 2        → online course, certificate, reflection
# ✓ CAT General  → exam, test, assessment, revision, marks
# ✓ Add Unit     → add unit, enroll, join
# ✓ Drop Unit    → drop unit, withdraw, remove
# ✓ Registration → register, fees, portal, timetable
# ✓ Exit         → bye, goodbye, exit, quit, done, thank
#
# NLP CONCEPTS DEMONSTRATED:
# ✓ Week 1 → Tokenization, stop words, stemming, lemmatization
# ✓ Week 2 → N-gram bigram model, POS tagging
# ✓ Week 3 → Sequence labeling (HMM concept)
# ✓ Week 4 → Dependency parsing, semantic similarity fallback
# ✓ Week 5 → Full integrated NLP pipeline


# ════════════════════════════════════════════════════════════════
# DELIVERABLE 4: REFLECTION ON CHALLENGES FACED
# ════════════════════════════════════════════════════════════════
#
# Challenge 1 — Case sensitivity breaking keyword matching
# The first challenge encountered was case sensitivity.
# When typing "Hello" with a capital H or "CAT 1" in uppercase,
# the chatbot returned the default "I did not understand"
# response instead of the correct one. This happened because
# Python string comparison is case sensitive — "Hello" and "hello"
# are treated as completely different strings.
# The solution was applying .lower() inside preprocess_input()
# before any matching runs, converting all input to lowercase first.
#
# Challenge 2 — 'cat' matching unrelated inputs
# Any sentence containing 'cat' (including 'I have a cat')
# would trigger the CAT schedule response incorrectly.
# Solution: checked for 'cat 1' and 'cat 2' as bigrams first
# using the Week 2 N-gram model before checking the single word
# 'cat', so specific inputs get precise responses.
#
# Challenge 3 — Empty input crashing the loop
# Pressing Enter without typing caused word_tokenize() to receive
# an empty string and produce an empty list, confusing the
# response function.
# Solution: added if not user_input.strip() check at the top
# of the loop to catch and handle empty inputs gracefully.
#
# Challenge 4 — Semantic similarity returning warnings
# spaCy's en_core_web_sm model showed a UserWarning about vectors
# when computing similarity because the small model has limited
# word vectors. Scores were lower than expected.
# Solution: used 0.75 as the minimum threshold for a semantic
# match and treated it as a fallback only when keyword matching
# completely failed.
# ════════════════════════════════════════════════════════════════
```
